```python
import numpy as np
import cv2
import mediapipe as mp

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    
    video = cv2.VideoCapture(0)
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = hands.process(image)
            
        orig_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        if results.multi_hand_landmarks:
            for hand, result in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(orig_image, result, mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec((0, 255, 0), 2, 2), 
                                          mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
                
    
        cv2.imshow('Webcam', orig_image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```

```python
def calculate_angle(a, b, c):
    
    angle = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(angle * 180 / np.pi)
    
    if angle > 180:
        angle = 360 - angle
        
    return angle
```

```python
x1=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x
y1=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y
x = np.array([x1, y1])
x2=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].x
y2=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].y
y = np.array([x2, y2])

x3=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_PIP].x
y3=results.landmark[mp_hands.HandLandmark.INDEX_FINGER_PIP].y
z = np.array([x3, y3])

calculate_angle(x,y,z)
```

```python
with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands:
    
    video = cv2.VideoCapture(0)
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = hands.process(image)
            
        orig_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        if results.multi_hand_landmarks:
            for hand, result in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(orig_image, result, mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec((0, 255, 0), 2, 2), 
                                          mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
                
        x1=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x
        y1=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y
        x = np.array([x1, y1])
        x2=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].x
        y2=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_DIP].y
        y = np.array([x2, y2])
        x3=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP].x
        y3=result.landmark[mp_hands.HandLandmark.INDEX_FINGER_MCP].y
        z = np.array([x3, y3])
        angle1 = calculate_angle(x,y,z)
        
        x5=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].x
        y5=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y
        xx = np.array([x5, y5])
        x6=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_DIP].x
        y6=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_DIP].y
        yy = np.array([x6, y6])
        x7=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP].x
        y7=result.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP].y
        zz = np.array([x7, y7])
        angle2 = calculate_angle(xx,yy,zz)
        
        x8=result.landmark[mp_hands.HandLandmark.RING_FINGER_TIP].x
        y8=result.landmark[mp_hands.HandLandmark.RING_FINGER_TIP].y
        xxx = np.array([x8, y8])
        x9=result.landmark[mp_hands.HandLandmark.RING_FINGER_DIP].x
        y9=result.landmark[mp_hands.HandLandmark.RING_FINGER_DIP].y
        yyy = np.array([x9, y9])
        x10=result.landmark[mp_hands.HandLandmark.RING_FINGER_MCP].x
        y10=result.landmark[mp_hands.HandLandmark.RING_FINGER_MCP].y
        zzz = np.array([x10, y10])
        angle3 = calculate_angle(xxx,yyy,zzz)
        
        x11=result.landmark[mp_hands.HandLandmark.PINKY_TIP].x
        y11=result.landmark[mp_hands.HandLandmark.PINKY_TIP].y
        xxxx = np.array([x1, y1])
        x12=result.landmark[mp_hands.HandLandmark.PINKY_DIP].x
        y12=result.landmark[mp_hands.HandLandmark.PINKY_DIP].y
        yyyy = np.array([x2, y2])
        x13=result.landmark[mp_hands.HandLandmark.PINKY_MCP].x
        y13=result.landmark[mp_hands.HandLandmark.PINKY_MCP].y
        zzzz = np.array([x3, y3])
        angle4 = calculate_angle(xxxx,yyyy,zzzz)
        
        cv2.putText(orig_image, str(np.round(angle1, 2)), tuple(np.multiply(np.array([x2, y2]), [640, 480]).astype(int)),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
        
        cv2.putText(orig_image, str(np.round(angle2, 2)), tuple(np.multiply(np.array([x6, y6]), [640, 480]).astype(int)),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
        
        cv2.putText(orig_image, str(np.round(angle3, 2)), tuple(np.multiply(np.array([x9, y9]), [640, 480]).astype(int)),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
        
        cv2.putText(orig_image, str(np.round(angle4, 2)), tuple(np.multiply(np.array([x12, y12]), [640, 480]).astype(int)),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
    
        cv2.imshow('Webcam', orig_image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```
