```python
import cv2
import mediapipe as mp
import numpy as np
```

```python
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
```

```python
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    
    video = cv2.VideoCapture('production_id_5195148 (1080p).mp4')
    
    while True:
        
        ret, frame = video.read()
        blue_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(blue_frame)
        
        orig_frame = cv2.cvtColor(blue_frame, cv2.COLOR_RGB2BGR)
        orig_frame = cv2.resize(orig_frame, (1000, 600))
        
        mp_drawing.draw_landmarks(orig_frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec((255, 0, 0), 2, 2),
                                 mp_drawing.DrawingSpec((255, 0, 0), 2, 2))
        
        cv2.imshow('frame', orig_frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    
    video.release()
    cv2.destroyAllWindows()
```

```python
x1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x
y1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y
shoulder = np.array([x1, y1])

x2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x
y2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y
elbow = np.array([x2, y2])

x3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x
y3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
wrist = np.array([x3, y3])
```

```python
def calculate_angle(a, b, c):
    
    angle = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(angle * 180 / np.pi)
    
    if angle > 180:
        
        return (360 - angle)
    
    return angle
```

```python
calculate_angle(shoulder, elbow, wrist)
```

```python
flag = 0
count = 0
down = False

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    
    video = cv2.VideoCapture('production_id_5195148 (1080p).mp4')
    
    while True:
        
        ret, frame = video.read()
        blue_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(blue_frame)
        
        orig_frame = cv2.cvtColor(blue_frame, cv2.COLOR_RGB2BGR)
        orig_frame = cv2.resize(orig_frame, (1000, 600))
        
        mp_drawing.draw_landmarks(orig_frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec((255, 0, 0), 2, 2),
                                 mp_drawing.DrawingSpec((255, 0, 0), 2, 2))
        
        x1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x
        y1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y
        shoulder = np.array([x1, y1])
        
        x2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x
        y2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y
        elbow = np.array([x2, y2])
        
        x3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x
        y3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
        wrist = np.array([x3, y3])
        
        angle = calculate_angle(shoulder, elbow, wrist)
        
        
        cv2.putText(orig_frame, str(angle), 
                    tuple(np.multiply(np.array([x2, y2]), [640, 480]).astype(int)), 
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1, (0, 255, 0), 2, cv2.LINE_AA)
        
        if angle < 80:
            down = True
            flag+=1
            if flag<=1:
                count+=1
                
        elif angle > 80:
            down = False
            flag = 0
            
        cv2.putText(orig_frame, str(count), 
                    (50, 50), 
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1, (0, 255, 0), 2, cv2.LINE_AA)
        print(count)
        
        cv2.imshow('frame', orig_frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    
    video.release()
    cv2.destroyAllWindows()
```
