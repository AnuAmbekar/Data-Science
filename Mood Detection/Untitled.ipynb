```python
import cv2
import mediapipe as mp
import numpy as np
import csv
from itertools import chain
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pickle
```

```python
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
```

```python
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    video = cv2.VideoCapture(0)
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = holistic.process(image)
                
        orig_vid = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        mp_drawing.draw_landmarks(orig_vid, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION, mp_drawing.DrawingSpec((255, 0, 0), 2, 2), mp_drawing.DrawingSpec((255, 0, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, mp_drawing.DrawingSpec((0, 0, 255), 2, 2), mp_drawing.DrawingSpec((0, 0, 255), 2, 2))
        
        
    
        
        cv2.imshow('frame', orig_vid)
        
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```

```python
num_coords = len(results.face_landmarks.landmark) + len(results.pose_landmarks.landmark)
num_coords
```

```python
landmarks = ['class']

for i in range(num_coords):
    landmarks += [f'x{i}', f'y{i}', f'z{i}']
```

```python
with open('coords.csv', 'w') as file:
    writer = csv.writer(file)
    
    writer.writerow(landmarks)
```

```python
df = pd.read_csv('coords.csv')

X = df.drop(columns='class')
y = df['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

```python
model = pickle.dumps(clf)
```

```python
loaded_model = pickle.loads(model) 
```

```python
loaded_model.predict(X_test) 
```

```python
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    
    video = cv2.VideoCapture(0)
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = holistic.process(image)
                
        orig_vid = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        mp_drawing.draw_landmarks(orig_vid, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION, mp_drawing.DrawingSpec((255, 0, 0), 2, 2), mp_drawing.DrawingSpec((255, 0, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, mp_drawing.DrawingSpec((0, 255, 0), 2, 2), mp_drawing.DrawingSpec((0, 255, 0), 2, 2))
        mp_drawing.draw_landmarks(orig_vid, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS, mp_drawing.DrawingSpec((0, 0, 255), 2, 2), mp_drawing.DrawingSpec((0, 0, 255), 2, 2))
        
        
        try:   
            face_coords = [(i.x, i.y, i.z) for i in results.face_landmarks.landmark]
            pose_coords = [(i.x, i.y, i.z) for i in results.pose_landmarks.landmark]
            all_coords = face_coords + pose_coords
            prediction = loaded_model.predict(np.array(list(chain(*(all_coords)))).reshape(1, -1))[0]
            x=results.face_landmarks.landmark[mp_holistic.PoseLandmark.RIGHT_EAR].x
            y=results.face_landmarks.landmark[mp_holistic.PoseLandmark.RIGHT_EAR].y
            coordinates = tuple(np.multiply(np.array([x,y]), [640, 480]).astype(int))
            cv2.putText(orig_vid, str(prediction), coordinates, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
        
        except:
            pass
        
        cv2.imshow('frame', orig_vid)
        
        
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```
