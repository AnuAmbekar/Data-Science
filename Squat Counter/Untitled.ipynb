```python
import cv2
import mediapipe as mp
import numpy as np

mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
```

```python
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    
    video = cv2.VideoCapture('production_id_4921641 (1080p).mp4')
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = pose.process(image)
        
        resized = cv2.resize(frame, (400, 700))
        cv2.imshow('frame', resized)
        
        
        if cv2.waitKey(15) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```

```python
x1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x
y1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y
hip = np.array([x1, y1])

x2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x
y2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y
knee = np.array([x2, y2])

x3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x
y3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y
ankle = np.array([x3, y3])
```

```python
def calculate_angle(a, b, c):
    
    angle = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    
    angle = np.abs(angle * 180 / np.pi)
    
    if angle > 180:
        
        return 360 - angle
    
    return angle
```

```python
calculate_angle(hip, knee, ankle)
```

```python
flag = 0
count = 0
down = False

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    
    video = cv2.VideoCapture('production_id_4921641 (1080p).mp4')
    
    while True:
        
        ret, frame = video.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        results = pose.process(image)
        
        orig_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        resized = cv2.resize(orig_image, (400, 700))
        
        
        mp_drawing.draw_landmarks(resized, results.pose_landmarks, mp_pose.POSE_CONNECTIONS, 
                                 mp_drawing.DrawingSpec((255, 0, 0), 1, 1),
                                 mp_drawing.DrawingSpec((255, 0, 0), 1, 1))
        
        x1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x
        y1=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y
        hip = np.array([x1, y1])
        
        x2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x
        y2=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y
        knee = np.array([x2, y2])
        
        x3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x
        y3=results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y
        ankle = np.array([x3, y3])
        
        angle = calculate_angle(hip, knee, ankle)
        
        cv2.putText(resized, str(np.round(angle, 2)), tuple(np.multiply(np.array([x2, y2]), [400, 700]).astype(int)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
        
        if angle < 45:
            flag += 1
            down = True
            if flag <= 1:
                count+=1
                
        elif angle > 45:
            down = False
            flag = 0
            
        cv2.putText(resized, str(count), (40, 85),
                cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 2, cv2.LINE_AA)
        
        print(count)
        
        cv2.imshow('frame', resized)
        
        if cv2.waitKey(15) & 0xFF == ord('q'):
            break
            
    video.release()
    cv2.destroyAllWindows()
```
